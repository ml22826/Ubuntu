```python
import pandas as pd

results = []
prev_maf = None
start_pos = None
end_pos = None

#iterating through rows of df
for _, row in df.iterrows():
    #new range if MAF changes
    if row['MAF'] != prev_maf:
        if start_pos is not None:
            #savethe previous range
            results.append({'Pos-start': start_pos, 'Pos-end': end_pos, 'MAF': prev_maf})
        start_pos = row['POS']
        end_pos = row['POS']
        prev_maf = row['MAF']
    else:
        #extend current range
        end_pos = row['POS']

#save the last range
if start_pos is not None:
    results.append({'Pos-start': start_pos, 'Pos-end': end_pos, 'MAF': prev_maf})

#conversion results to a DataFrame
summary_df = pd.DataFrame(results)

#output the summary to a CSV file
summary_df.to_csv('maf_ranges_summary.csv', index=False)
```
